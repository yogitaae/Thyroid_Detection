```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = open("thyroid.csv")
df = pd.read_csv(file)
df
```

```python
df.drop("other",axis=1,inplace=True)
```

```python
feature_cols = ["age",
                "sex",
                "on_thyroxine",
                "query_on_thyroxine",
                "on_antithyroid_medication",
                "sick",
                "pregnant",
                "thyroid_surgery",
                "I131_treatment",
                "query_hypothyroid",
                "query_hyperthyroid",
                "lithium",
                "goitre",
                "tumor",
                "hypopituitary",
                "psych",
                "TSH measured",
                "TSH",
                "T3_measured",
                "T3",
                "TT4_measured",
                "TT4",
                "T4U_measured",
                "T4U",
                "FTI_measured",
                "FTI",
                "TBG_measured",
                "TBG",
               "target"]
```

```python
df.columns = feature_cols
df
```

```python
target = df.target
create = target.str.split('([A-Za-z]+)', expand=True)
create = create[1]
target = create.replace({None:'Z'})
df.target = target

df.target.unique()
```

```python
df
```

```python
df = df.replace(['?'],np.nan)
df.isnull().sum()
```

```python
df.drop(['TBG_measured','TBG','T3_measured','TSH measured','TT4_measured','T4U_measured','FTI_measured'],axis=1,inplace=True)

df
```

```python
df.sex.replace({'F':2,'M':1},inplace=True)
round_Values = round(df.sex.mean())
df.sex.fillna(round_Values,inplace=True)
df.sex.unique()
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import KNNImputer
knnimp = KNNImputer(n_neighbors=3)
```

```python
cols = ['TSH','T3','TT4','T4U','FTI']
for i in cols:
    df[i] = knnimp.fit_transform(df[[i]])
```

```python
X = df.drop('target',axis=1)
y = df.target
```

```python
y.unique()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
```

```python
X_pca = pd.DataFrame(data = v, columns = ['component_1', 'component_2', 'component_3', 'component_4', 'component_5', 'component_6', 'component_7', 'component_8', 'component_9', 'component_10'])
```

```python
if 'T4' in df.columns and 'T3' in df.columns:
    df['T3_T4_ratio'] = df['T3'] / df['T4']
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
rf_classifier = RandomForestClassifier(random_state=42)
```

```python
rf_classifier.fit(X_train, y_train)
```

```python
y_pred_rf = rf_classifier.predict(X_test)
```

```python
accuracy_rf = accuracy_score(y_test, y_pred_rf)
```

```python
print("Random Forest Accuracy:", accuracy_rf)
```
